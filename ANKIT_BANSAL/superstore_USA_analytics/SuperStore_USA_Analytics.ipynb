```python
#import python libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel("Superstore_USA.xlsx")
df
```

```python
df.info()
df.isna().sum()
```

```python
df['Product Base Margin'].fillna(df['Product Base Margin'].mean(),inplace=True)
```

```python
#univariate analysis
#count the number of order priority of all orders
ax = sns.countplot(data=df,x='Order Priority')
ax.bar_label(ax.containers[0])
plt.savefig("Order Priority Distribution.jpg")
plt.title("Order Priority Distribution")
```

```python
x = df['Ship Mode'].value_counts().index
y = df['Ship Mode'].value_counts().values
plt.figure(figsize=(7,4))
plt.pie(y,labels=x,startangle=180,autopct="%0.2f%%")
plt.legend(loc=1)

```

```python
#Bivariate Analysis
#number of products delivered by various mode for each product category
plt.figure(figsize=(4,3))
sns.countplot(data=df,x="Ship Mode",hue="Product Category")
```

```python

#number of products delivered by various mode for each Customer Segment
plt.figure(figsize=(6,3))
ax = sns.countplot(data=df,x="Customer Segment")
ax.bar_label(ax.containers[0])
```

```python
#number of products delivered by various mode for each product category
plt.figure(figsize=(8,3))
ax = sns.countplot(data=df[df["Product Category"]=="Office Supplies"],x="Product Category",hue="Product Sub-Category")
ax.bar_label(ax.containers[0])
```
