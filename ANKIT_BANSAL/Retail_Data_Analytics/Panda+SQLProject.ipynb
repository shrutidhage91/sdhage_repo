```python
import pandas as pd
#read file and replace non available values as Nan
orders=pd.read_csv("orders.csv",na_values=['Not Available', 'unknown'])
#display(orders)
#orders.head(20)
orders['Ship Mode'].unique()
```

```python
#convert column names to lower case and rerplace spaces by underscore
orders.columns = orders.columns.str.lower()
orders.columns = orders.columns.str.replace(' ','_')
orders.columns
orders
```

```python
#derive new columns
orders['discount'] = orders['list_price']*orders['discount_percent']*0.01

orders['selling_price'] = orders['list_price']-orders['discount']
orders['profit'] = orders['selling_price']-orders['cost_price']
orders.head(5)
```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'],format='%Y-%m-%d')
orders.dtypes
#orders
```

```python
orders.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
orders
```

```python
#find top 10 highest revenue generating products
query = "SELECT product_id,sum(selling_price) as sales_price FROM orders group by product_id order by sales_price  desc limit 10 "
#query = "select * from orders"

# Run the query
df_sql = psql.sqldf(query, locals())
print(df_sql)
```

```python
#find top 5 highest selling product in each region.
query = f"""select * from 
(select region,product_id,sales,dense_rank() over (partition by region order by sales desc) as rnk from  
(SELECT region,product_id,sum(selling_price)  as sales from orders group by product_id,region)) where rnk<=5"""
#or with cte we can do
query ="with cte as (SELECT region,product_id,sum(selling_price)  as sales from orders group by product_id,region) select * from ( select *,dense_rank() over (partition by region order by sales desc) as rnk  from cte) where rnk<=5"

# Run the query
df_sql = psql.sqldf(query, locals())
print(df_sql)
```

```python
# find month over month growth over sales eg jan 2022 vs jan2023
query ="with cte as (select STRFTIME('%m', order_date)  as month,STRFTIME('%Y',order_date) as year, sum(selling_price) as sales from orders  where STRFTIME('%Y',order_date) in ('2022','2023') group by STRFTIME('%m', order_date),STRFTIME('%Y', order_date) ) select T22.month,T22.sales as sales_2022 , T23.sales as sales_2023 from cte T22 inner join cte T23 on T22.month=T23.month and T22.year<>T23.year where T22.year='2022' "
# Run the query
df_sql = psql.sqldf(query, locals())
print(df_sql)
```

```python
#for each category which month  had highest sales?
query =f"""with cte as (select STRFTIME('%Y%m', order_date)  as month,
category, sum(selling_price) as sales from orders    group by 
category,STRFTIME('%Y%m', order_date) )
select * from (select *, row_number() over (partition by category order by sales desc) as rn
from cte) where rn=1
 """

# Run the query
temp2 = psql.sqldf(query, locals())
print(temp2)
```
