```python
#which coutry has highest comedy movies
query = """select count(nl.show_id) as cnt_comendy,nc.country
from nf_v2 nf 
inner join nf_listed_in nl 
on nf.show_id=nl.show_id
inner join nf_country nc 
on nc.show_id=nl.show_id
 where nl.listed_in ='Comedies' and nf.type='Movie' group by nc.country order by count(nl.listed_in) desc limit 1 """

df_sql = psql.sqldf(query,locals())
df_sql
```

```python
#for each year as per date added which director have highest number of movies released on netflix
query ="""with cte as (select nd.director,STRFTIME('%Y',nf.date_added)  as year,count(nf.show_id),
dense_rank() over (partition by STRFTIME('%Y',nf.date_added) order by count(nf.show_id) desc,nd.director desc) as rnk
from nf_v2 nf
inner join nf_director nd
on nf.show_id=nd.show_id where nd.director is not null and nf.type='Movie' group by nd.director,STRFTIME('%Y',nf.date_added) )
select * from cte where rnk=1 order by year desc"""
df_sql = psql.sqldf(query,locals())
df_sql
```
