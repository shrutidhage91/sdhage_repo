```python
#import python libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding= 'unicode_escape')
df
df.shape
df.info()
```

```python
df.drop(["Status","unnamed1"], axis=1, inplace=True)
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
#change amount data tyoe
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.rename(columns={"Marital_Status":"shadi"})
```

```python
df
```

```python
#bar chart for gender
plt.figure(figsize=(3,4))
plt.title("Gender distribution")
ax = sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#plot a bar chart for gender vs total amount
df2 = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
#sns.barplot(data=df2,x="Gender",y="Amount")
sns.barplot(x = 'Gender',y= 'Amount' ,data = df2)
```

```python
ax = sns.countplot(data=df,x='Age Group',hue='Zone')
```

```python
#age group vs amount
df3 = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=df3)
```

```python
#total number of orders from top 10 states
plt.figure(figsize=(12,4))
df4 = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x="State",y="Orders",data=df4)
```

```python
#total amount from top 10 states
df5=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
plt.figure(figsize=(12,4))
sns.barplot(x="State",y="Amount",data=df5)
```

```python
#what is the count of married vs unmarried people who purchased
ax=sns.countplot(data=df,x="Marital_Status")
ax.bar_label(ax.containers[0])
```

```python
#amount spent by married vs unmarried peopleof each gender
df6 = df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
plt.figure(figsize=(5,8))
sns.barplot(data=df6,x="Marital_Status",y="Amount",hue="Gender")
df6
```

```python
#number of buyers from every occupation
plt.figure(figsize=(12,4))
sns.countplot(data=df,x="Occupation")
```

```python
df7 = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
plt.figure(figsize=(12,4))
sns.barplot(x="Occupation",y="Amount",data=df7)
```

```python
#number of products sold from product_category 
plt.figure(figsize=(22,4))
sns.countplot(data=df,x="Product_Category")
```

```python
df8 = df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
plt.figure(figsize=(22,4))
sns.barplot(x="Product_Category",y="Amount",data=df8)
```

```python
#Number of orders placed for every product
df9 = df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
plt.figure(figsize=(22,4))
sns.barplot(x="Product_ID",y="Orders",data=df9)
```
